# MODEL FITTING

```python
# Splitting into train and test for calculating the accuracy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

## Support Vector Machine 

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='rbf',C=15,gamma=0.001,random_state=100)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict_svm = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_predict_svm)
print(accuracy_score(y_test,y_predict_svm))
```

```python
print(classification_report(y_test,y_predict_svm))
```

```python
pd.crosstab(y_test,y_predict_svm)
```

## Decision Tree

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=5,random_state=15)
model.fit(X_train,y_train)
```

```python
#Visualizing
dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, 
                feature_names = X.columns,
                
                rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
y_predict_dt = model.predict(X_test)
print(accuracy_score(y_test, y_predict_dt))
```

```python
pd.crosstab(y_test,y_predict_dt)
```

# Decision tree with grid search

```python
from sklearn.model_selection import GridSearchCV
classifier=DecisionTreeClassifier(random_state=50,splitter='best',max_depth=6)
parameter_1=[{'min_samples_split':[2,3,4,5],'criterion':['entropy']}]

model_grid=GridSearchCV(estimator=classifier, param_grid=parameter_1, scoring='accuracy',cv=10)
model_grid.fit(X_train,y_train)
```

```python
y_predict_grid = model_grid.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict_grid))
print(classification_report(y_test,y_predict_grid))
```

```python
pd.crosstab(y_test,y_predict_grid)
```
