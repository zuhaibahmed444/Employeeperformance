## Outliers imputation 

```python
data.loc[data.TotalWorkExperienceInYears>38,'TotalWorkExperienceInYears'] =np.nan
data.fillna(data.mean(),inplace=True)

data.loc[data.ExperienceYearsAtThisCompany>35,'ExperienceYearsAtThisCompany'] = np.nan
data.fillna(data.mean(), inplace=True)

data.loc[data.YearsWithCurrManager>16,'YearsWithCurrManager'] = np.nan
data.fillna(data.mean(), inplace=True)

data.loc[data.YearsSinceLastPromotion>12,'YearsSinceLastPromotion'] = np.nan
data.fillna(data.mean(), inplace=True)

data.loc[data.ExperienceYearsInCurrentRole>15,'ExperienceYearsInCurrentRole'] = np.nan
data.fillna(data.mean(), inplace=True)
```

# Top features which Impact on Performance of the Employee

- to select the best features we will use select best features SKlearn package i.e SelectKBest

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
```

```python
#let us encode the data to fit the model 
lis = [2,3,4,5,6,7,16,26]
enc = LabelEncoder()
for i in lis:
    data.iloc[:,i] = enc.fit_transform(data.iloc[:,i])
data.head()
```

```python
X = data.iloc[:,1:-1]  #feature columns we drop employee id 
y = data.iloc[:,-1]    #target column i.e Performance rating
```

### univariate selection 

```python
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featuresc = pd.concat([dfcolumns,dfscores],axis=1)
featuresc.columns = ['Specs','Score']  #naming the dataframe columns
print(featuresc.nlargest(10,'Score'))  #print 10 best features
```

### Feature improtance

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
feture= pd.Series(model.feature_importances_, index=X.columns)
feture.plot(kind='bar')
plt.show()
```

```python
feture.nlargest(10).plot(kind='bar')
```

## Conclusion for feature importnace 
### Top features which affect the performance of the employee are
- EmpEnvironmentSatisfaction
- EmpLastSalaryHikePercentage
- YearsSinceLastPromotion

# MODEL FITTING

```python
# Splitting into train and test for calculating the accuracy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

## Support Vector Machine 

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='rbf',C=15,gamma=0.001,random_state=100)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict_svm = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_predict_svm)
print(accuracy_score(y_test,y_predict_svm))
```

```python
print(classification_report(y_test,y_predict_svm))
```

```python
pd.crosstab(y_test,y_predict_svm)
```

## Decision Tree

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=5,random_state=15)
model.fit(X_train,y_train)
```

```python
#Visualizing
dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, 
                feature_names = X.columns,
                
                rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
y_predict_dt = model.predict(X_test)
print(accuracy_score(y_test, y_predict_dt))
```

```python
pd.crosstab(y_test,y_predict_dt)
```

# Decision tree with grid search

```python
from sklearn.model_selection import GridSearchCV
classifier=DecisionTreeClassifier(random_state=50,splitter='best',max_depth=6)
parameter_1=[{'min_samples_split':[2,3,4,5],'criterion':['entropy']}]

model_grid=GridSearchCV(estimator=classifier, param_grid=parameter_1, scoring='accuracy',cv=10)
model_grid.fit(X_train,y_train)
```

```python
y_predict_grid = model_grid.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict_grid))
print(classification_report(y_test,y_predict_grid))
```

```python
pd.crosstab(y_test,y_predict_grid)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=28)
model.fit(X_train, y_train)
y_predict_kn = model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_predict_kn))
pd.crosstab(y_test, y_predict_kn)
```

## Random Forest with gridsearch

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators =30,max_depth =5,max_features = 26,criterion='gini',random_state=150)
```

```python
parameters=[{'min_samples_split':[2,3,4,5],'criterion':['gini']}]

model_grf=GridSearchCV(estimator=model_rf, param_grid=parameters, scoring='accuracy',cv=10)
model_grf.fit(X_train,y_train)
```

```python
y_predict_grf = model_grf.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict_grf))
print(classification_report(y_test,y_predict_grf))
pd.crosstab(y_test,y_predict_grf)
```
